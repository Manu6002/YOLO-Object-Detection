```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x,y = iris.data,iris.target
```

```python
print("feature name (what x column names are):",iris.feature_names)
```

```python
print("feature name (what y column names are):",iris.target_names)
```

```python
import pandas as pd
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size=0.35,random_state=62,stratify=y)
```

```python
clf = LogisticRegression(max_iter = 200,multi_class ="multinomial")
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
print(f"accuracy = {accuracy:.4f}")
```
